# [1] Instrucciónes:
 **Genere una señal ruidosa. Esta señal debe contener al menos dos frecuencias principales distintas y ruido. Adicionalmente se puede usar como señal de entrada un audio corto que contenga ruido. Grafique la señal generada. No olvide indicar las etiquetas de los ejes y el título del gráfico.**

##Pseudocódigo:
1. Definir los parámetros del problema (frecuencias, amplitudes, intervalo de tiempo etc).
2. Crear el array con los valores de tiempo donde se van a evaluar las señales.
3. Crear una señal sin rudio (suma de senos determinados por los parámetros iniciales.)
4. Utilizando la señal previa se crea una nueva señal donde se introduce ruido de manera aleatoria. El ruido sigue una distribución normal cuyos parámetros fueron definidos previamente.
5. Visulizar las dos señales en el intervalo de tiempo establecido.  

```python
import numpy as np
import pandas as pd
import scipy.fft as ft
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
"""
Parámetros
"""
t0 = 0
tf = 1
N = 200
l = int(N/2) #valor en el que se reducen intervalos para apreciar mejor los resultados de la FfT

f0 = 5
a0 = 32
f1 = 12
a1 = 23
f2 = 9
a2 = 17

media = 0
sigma = 50
umbral = 16
```

```python
"""
Creación de la señal con y sin ruido.
"""
t = np.linspace(t0,tf,num=N) # tiempo
y_pura = a0*np.sin(2*np.pi*f0*t)+a1*np.sin(2*np.pi*f1*t)+a2*np.sin(2*np.pi*f2*t) # señal sin ruido 
y_ruido = a0*np.sin(2*np.pi*f0*t)+a1*np.sin(2*np.pi*f1*t)+a2*np.sin(2*np.pi*f2*t)+np.random.normal(loc=media,scale=sigma,size=N) # señal con ruido 
```

```python
"""
Creación de gráficas con la librería plotly.
"""
fig = make_subplots(rows=2, cols=1, subplot_titles=("Señal","Señal Ruidosa"))

fig.add_trace(go.Scatter(x=t, y=y_pura),row=1, col=1)
fig.add_trace(go.Scatter(x=t, y=y_ruido),row=2, col=1)

fig.update_xaxes(title_text="Tiempo", row=1, col=1)
fig.update_xaxes(title_text="Tiempo", row=2, col=1)
fig.update_yaxes(title_text="Amplitud", row=1, col=1)
fig.update_yaxes(title_text="Amplitud", row=2, col=1)

fig.update_layout(height=700, width=900, title_text="Señales de entrada:",
                  showlegend=False)

fig.show()
```

## [1] Análisis y Comentarios:
En la primera gráfica podemos apreciar la señal "pura" o sin ruido, la cual tiene la forma:
$$
y(t)_{pura} = a_{0}sin(2\pi f_{0}t)+a_{1}sin(2\pi f_{1}t)+a_{2}sin(2\pi f_{2}t)
$$
En la segunda gráfica tenemos la señal con ruido la cual sigue:
$$
y(t,\mu,\sigma,N)_{ruido}= y(t)_{pura}+\xi (\mu ,\sigma,N)=y_{ruido}
$$
Donde:
> $\xi (\mu ,\sigma,N)$: es una función que genera números aleatorios siguiendo la distribución normal.

> $\mu$: representa la media de distribución.
 
> $\sigma$: determina la desviación estandar de la distribución.

> $N$: señaliza la cantidad de puntos que se utilizaron para generar cada señal. 

**Cabe destacar** que cada vez que se genera $y_{ruido}$ se obtienen señales distintas debido a la naturaleza aleatoria de $\xi$ por lo que los resultados finales serán distintos cada vez.

# [2]Instrucciónes:
**Calcule la transformada de Fourier de la señal de entrada. Grafique esta señal en el dominio de la frecuencia.**

##Pseudocódigo:

1. Utilizando la función fft de la librería scipy.fft se calcula la trasformada rápida de Fourier para $y_{pura}$ y $y_{ruido}$
2. Para cada una de las fft se calcula su magnitud estandarizada (dictamina cuanto de una frecuencia pura hay en una señal dada.)
3. Se genera un array que va a representar las frecuencias obtenidas.
4. Se visualizan los datos obtenidos. Se puede determinar la longitud del intervalo a mostrar con el parámetro $l$.

```python
"""
Obtención de FFT y su magnitud para ambas señales.
"""
fft_pura = ft.fft(y_pura)
fft_ruido = ft.fft(y_ruido)
magnitud_fft_pura = (2/y_pura.size)*abs(ft.fft(y_pura)) #valor normalizado de la magnitud de la FFT
magnitud_fft_ruido = (2/y_ruido.size)*abs(ft.fft(y_ruido)) #valor normalizado de la magnitud de la FFT
frec = np.arange(0,len(y_pura))
```

```python
"""
Gráficas de las señales en el dominio de la frecuencia.
"""
fig = make_subplots(rows=2, cols=1, subplot_titles=("Señal en el dominio de la frecuencia","Señal ruidosa en el dominio de la frecuencia"))

fig.add_trace(go.Scatter(x=frec[:l], y=magnitud_fft_pura[:l]),row=1, col=1) #se reducen los intervalos para poder apreciar los resultados
#fig.add_trace(go.Scatter(x=frec[:l], y=magnitud_fft_ruido[:l]),row=2, col=1) #se reducen los intervalos para poder apreciar los resultados
fig.add_trace(go.Scatter(x=frec, y=magnitud_fft_ruido),row=2, col=1) #vista total

fig.update_xaxes(title_text="Frecuencia [Hz]", row=1, col=1)
fig.update_xaxes(title_text="Frecuencia [Hz]", row=2, col=1)
fig.update_yaxes(title_text="Magnitud de FFT", row=1, col=1)
fig.update_yaxes(title_text="Magnitud de FFT", row=2, col=1)

fig.update_layout(height=700, width=900, title_text="Señales en el dominio de la frecuencia:",
                  showlegend=False)

fig.show()
```

## [2]Análisis y Comentarios:

Hay bastantes observaciones que hacer con respecto a las gráficas anteriores:

> Si observamos con detenimiento podemos notar que los intervalos de frecuencia para ambas señales es distinto.

Esto se puede modificar a gusto en el código y se debe a la aleatoriedad de la señal ruidosa. 

Debido a que en cada iteración del código se introduce ruido de manea aletoria resulta una necesidad visualizar por completo los resultados de $y_{ruido}$, por el contrario no hay necesidad de visualizar por completo los resultados de $y_{pura}$ dado a que se sabe previamente el lugar donde se encontraran las frecuencias.

> Las  amplitudes de cada frecuencia son los valores normalizados de la magnitud de la transformada rápida de Fourier.

De la teroría de la transdormada de Fourier sabemos que:
$$\hat{f}(\nu)=\int_{-\infty }^{\infty }f(t)e^{-2\pi i \nu t }dt$$
$$\hat{f}(\nu)=\int_{-\infty }^{\infty }f(t)cos(2\pi\nu t)-i\int_{-\infty }^{\infty }f(t)sin(2\pi\nu t) = a-bi$$
$$\left| \hat{f} \right|=\sqrt{a^{2}+b^{2}}$$

 Donde el valor absoluto de $\hat{f}$ es la magnitud de un número complejo, que representa cuanto de una frecuencia hay en una señal dada.

 > Al agragar valores aleatorios a $y_{pura}$ conseguimos con exito agregar frecuencias no deseadad a la señal.

 Cabe destacar que la gran cantidad frecuencias no deseadas y su amplitud, es distinta cada vez que se corre el código.

# [3]Instrucciónes:
**Escriba una función llamada Filtrar_Señal() que tome la señal en el dominio de
la frecuencia y filtre el ruido. Esta función tiene que tener como parámetros de
entrada, tanto la señal en el dominio de la frecuencia como un parámetro llamado umbral . Este parámetro indica una amplitud de referencia para el filtrado de
la señal. La función debe retornar la señal filtrada, siempre en el dominio de la frecuencia. Grafique la señal filtrada en el dominio de la frecuencia.**

##Pseudocódigo:
1. De la gráfica anterior se determina un umbral.
2. A partir del umbral se crea una función para filtrar la señal.
3. Dicha función recive la señal con ruido y evalua cada punto en el plano complejo. Si la magnitud estandar del punto es menor que el valor umbral se cambia el punto por un 0 eliminando así cualquier frecuencia por debajo del valor umbral.
4. Se visualizan los datos de la señal.

```python
def Filtrar_Señal(y,umbral):
  """
  Función que permite eliminar todas las frecuencia 
  """
  for i in range(y.size):
    a = (2/y.size)*abs(y[i])
    if a <= umbral:
      y[i] = 0
  return y
```

```python
"""
Filtrado de la señal
"""
fft_ruido_filtrada = Filtrar_Señal(fft_ruido,umbral)
magnitud_fft_ruido_filtrada =  (2/y_ruido.size)*abs(fft_ruido_filtrada) #valor normalizado de la magnitud de la FFT
```

```python
"""
Proceso de graficación con plotly express
"""
df = pd.DataFrame({"Magnitud de FFT":magnitud_fft_ruido_filtrada[:l],"frecuencia[Hz]":frec[:l]})
px.line(df,x="frecuencia[Hz]",y="Magnitud de FFT",title="Magnitud de la FFT de la señal con ruido filtrada en el dominio de la frecuencia.") #se reducen los intervalos para poder apreciar los resultados
```

## [3]Análisis y Comentarios:

Aunque la gráfica anterior es casi identica a la obtenida por la señal pura es importante destacar la importancia del parámetro umbral.

Dada la naturaleza aleatoria del ruido, en general se tiene que entre mayor sea los valores  $\mu$ y $\sigma$ mayor será el ruido. Este ruido se refleja en una mayor cantidad de frecuencias no desadas y en amplitudes más grandes de las mismas. Por lo que se tienen que escoger, en promedio, valores más altos de umbral entre más grandes sean los valores de  $\mu$ y $\sigma$, lo contrario tan bien aplica.

De esta forma hay  una relación directa entre el ruido y el umbral y resulta necesario escoger valores adecuados de umbral para filtrar bien la señal.


# [4]Instrucciónes:
**Finalmente, aplique la transformada rápida inversa a la señal filtrada y grafique la señal resultante.**

##Pseudocódigo:

1. Por medio de scipy.fft se calcula la tranformada rápida inversa de la señal filtrada.
2. A partir de los resultados anteriores se visualizan los datos finales.

```python
"""
Cálculo de la inversa de FFT de la señal de ruido filtrada.
"""
inversa_fft_ruido_filtrada = ft.ifft(fft_ruido_filtrada)
```

```python
"""
Creación de gráficas con la librería plotly.
"""
fig = make_subplots(rows=2, cols=2,subplot_titles=("Señal Pura","Señal Ruidosa","Señal Ruidosa Filtrada (Parte Real)","Señal Ruidosa Filtrada (Parte Imaginaria)"))

fig.add_trace(go.Scatter(x=t, y=y_pura),row=1, col=1)
fig.add_trace(go.Scatter(x=t, y=y_ruido),row=1, col=2)
fig.add_trace(go.Scatter(x=t, y=inversa_fft_ruido_filtrada.real),row=2, col=1)
fig.add_trace(go.Scatter(x=t, y=inversa_fft_ruido_filtrada.imag),row=2, col=2)


fig.update_xaxes(title_text="Tiempo", row=1, col=1)
fig.update_xaxes(title_text="Tiempo", row=1, col=2)
fig.update_xaxes(title_text="Tiempo", row=2, col=1)
fig.update_xaxes(title_text="Tiempo", row=2, col=2)
fig.update_yaxes(title_text="Amplitud", row=1, col=1)
fig.update_yaxes(title_text="Amplitud", row=1, col=2)
fig.update_yaxes(title_text="Amplitud", row=2, col=1)
fig.update_yaxes(title_text="Amplitud", row=2, col=2)

fig.update_layout(height=700, width=900, title_text="Señal original, con ruido y filtrada",
                  showlegend=False)

fig.show()
```
